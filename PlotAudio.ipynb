```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import sys
%matplotlib notebook
import glob
import os
import time

fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

def get_latest_file():
    list_of_files = glob.glob('/Users/pranavlal/Documents/Big_Data/Project/dhishoom/DetectedGunshots/*') # * means all if need specific format then *.csv
    latest_file = max(list_of_files, key=os.path.getctime)
    code = 0
    if "gun" in latest_file:
        code = 1
    return code, latest_file

while 1:
    ax.clear()
    code, latest_file = get_latest_file()
    sr, wav_data = wavfile.read(latest_file)
    Time=np.linspace(0, len(wav_data)/sr, num=len(wav_data))
    if code:
        plt.title("GUNSHOT DETECTED!!")
        ax.plot(Time, wav_data, 'r-')
    else:
        plt.title("All Good")
        ax.plot(Time, wav_data, 'b-')
    
    fig.canvas.draw()
    time.sleep(10)
                                
```
